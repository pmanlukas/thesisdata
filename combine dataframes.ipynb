# combine the dataframes
In this notebook we gonna combine the dataframes into one common dataframe, that lists the clusters from the used cluster approaches
The aim is to inspect the clustered data

```python
import pickle
import json
import pandas as pd
import numpy as np
```

```python
fieldsdf = pd.DataFrame()

with open('obj/onehotdf0518.pkl', 'rb') as handle:
    fieldsdf = pickle.load(handle)

fieldsdf.head()

```

```python
infodf = pd.DataFrame()

with open('obj/infohotdf0518.pkl', 'rb') as handle:
    infodf = pickle.load(handle)

infodf.head()
```

```python
keys_field = list(fieldsdf.keys())
keys_field
```

```python
keys_info = list(infodf.keys())
keys_info
```

```python
type(infodf)
```

```python
type(fieldsdf)
```

```python
comp_df = pd.concat([fieldsdf, infodf], axis=1)
```

```python
comp_df.head()
```

```python
comp_df.to_csv('comp_df0518.csv', encoding='utf8')
```
