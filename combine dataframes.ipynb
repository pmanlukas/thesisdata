# combine the dataframes
In this notebook we gonna combine the dataframes into one common dataframe, that lists the clusters from the used cluster approaches
The aim is to inspect the clustered data

```python
import pickle
import json
import pandas as pd
import numpy as np
```

```python
fieldsdf = pd.DataFrame()

with open('obj/onehotdf.pkl', 'rb') as handle:
    fieldsdf = pickle.load(handle)

fieldsdf.head()

```

```python
infodf = pd.DataFrame()

with open('obj/infohotdf.pkl', 'rb') as handle:
    infodf = pickle.load(handle)

infodf.head()
```

```python
keys_field = list(fieldsdf.keys())
keys_field[:-2]
```

```python
field_df = pd.DataFrame(fieldsdf[keys_field[:-2]])
```

```python
field_df.head()
```

```python
keys_info = list(infodf.keys())
keys_info[:-2]
```

```python
info_df = pd.DataFrame(infodf[keys_info[:-2]])
```

```python
info_df.head()
```

```python
info_df = pd.DataFrame(info_df)
```

```python
field_df = pd.DataFrame(field_df)
```

```python
type(info_df)
```

```python
type(field_df)
```

```python
comp_df = pd.concat([field_df, info_df], axis=1)
```

```python
comp_df.head()
```

```python
comp_df.to_csv('comp_df.csv', encoding='utf8')
```

```python
with open('obj/'+ 'comp_df' + '.pkl', 'wb') as f:
        pickle.dump(comp_df, f, pickle.HIGHEST_PROTOCOL)
```
