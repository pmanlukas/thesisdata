```python
import pickle
import json
import pandas
import scipy
```

```python
clust_dict = dict()

with open('obj/textclust_dict.pkl', 'rb') as handle:
    clust_dict = pickle.load(handle)

keys = list(clust_dict.keys())
print(keys)
```

```python
clust_dict[0]
```

```python
sentence_dict = dict()
```

```python
for cluster in clust_dict:
    sentence = clust_dict[cluster]
    sen = ' '.join(sentence)
    sentence_dict[cluster] = sen
```

```python
sentence_dict[1]
```

```python
keywords_dict = dict()
```

```python
liteClient.getKeywords
```

```python
for sent in sentence_dict:
    sentence = sentence_dict[sent]
    keywords = liteClient.getKeywords(sentence)
    keywords_dict[sent] = keywords
```

```python
for keyword in keywords_dict:
    print(keywords_dict[keyword])
```

```python
for cluster in clust_dict:
    print(clust_dict[cluster])
```

```python
reduce_keys = dict()
```

```python
for keys in keywords_dict:
    sentence = keywords_dict[keys]
    sen = ' '.join(sentence)
    keywords = liteClient.getKeywords(sen)
    reduce_keys[keys] = keywords
```

```python
for key in reduce_keys:
    print(reduce_keys[key])
```

```python
with open('obj/'+ 'reduce_keys' + '.pkl', 'wb') as f:
        pickle.dump(reduce_keys, f, pickle.HIGHEST_PROTOCOL)
```

```python
with open('obj/'+ 'keywords_dict' + '.pkl', 'wb') as f:
        pickle.dump(keywords_dict, f, pickle.HIGHEST_PROTOCOL)
```

```python
cluster_fingerprints = dict()
```

```python
for sent in sentence_dict:
    sentence = sentence_dict[sent]
    fingerprint = liteClient.getFingerprint(sentence)
    cluster_fingerprints[sent] = fingerprint
    
```

```python
with open('obj/'+ 'cluster_fingerprints' + '.pkl', 'wb') as f:
        pickle.dump(cluster_fingerprints, f, pickle.HIGHEST_PROTOCOL)
```
