```python
def pickleobj(name, obj):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
    print("pickled {}".format(name))
```

```python
versions = {}

with open('obj/version_dict.pkl', 'rb') as handle:
    versions = pickle.load(handle)
```

```python
def getjson(url):
    req = requests.get(url)
    specification = req.json()
    return specification
```

```python
def getspec(dicts):
    specdict = {}
    for url in dicts:
        spec = getjson(dicts[url])
        specdict[url] = spec
    print("success")
    return specdict
```

```python
entries = requests.get('https://api.apis.guru/v2/list.json')
```

```python
entries = entries.json()
```

```python
swagger_urls = {}
for ent in entries:
    entry = entries[ent]
    version = versions[ent][:]
    version = version[0]
    url_entry = entry['versions'][version]['swaggerUrl']
    swagger_urls[ent] = url_entry

print(len(swagger_urls.keys()))
```

```python
diff_specs = getspec(swagger_urls)
```
