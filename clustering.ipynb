# K-modes clustring
here we use k-modes clustering to cluster the apis based on their structure

```python
# random categorical data
data = clusterdf

km = KModes(n_clusters=15, init='Huang', n_init=5, verbose=3)
clusters = km.fit_predict(data)
# Print the cluster centroids
print(km.cluster_centroids_)
```

```python
import numpy as np
import pandas as pd
from kmodes import kmodes

df_dummy = pd.get_dummies(clusterdf)

#transform into numpy array
x = df_dummy.reset_index().values

km = kmodes.KModes(n_clusters=5, init='Huang', n_init=5, verbose=0)
clusters = km.fit_predict(x)
df_dummy['clusters'] = clusters


import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
pca = PCA(2)

# Turn the dummified df into two columns with PCA
plot_columns = pca.fit_transform(df_dummy.ix[:,0:15])

# Plot based on the two dimensions, and shade by cluster label
plt.scatter(x=plot_columns[:,1], y=plot_columns[:,0], c=df_dummy["clusters"], s=60)
plt.savefig('kmodes.pdf')
plt.savefig('kmodes.png')
plt.show()
```

```python
def createcluster(data, num_cluster, runs=10, level=2):
    km = KModes(n_clusters=num_cluster, init='Huang', n_init=runs, verbose=level)
    clusters = km.fit_predict(data)
    print(km.cluster_centroids_)
    print(km.labels_)
    return clusters
```

```python
cluster_dict = dict()
```

```python
for i in range(5,13):
    cluster_run = createcluster(data, i)
    cluster_dict[i] = cluster_run
```
