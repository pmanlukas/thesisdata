# K-modes clustring
here we use k-modes clustering to cluster the apis based on their structure

```python
# random categorical data
data = onehotdf

km = KModes(n_clusters=8, init='Huang', n_init=10, verbose=3)
clusters = km.fit_predict(data)

# Print the cluster centroids
print(km.cluster_centroids_)
```

```python
def createcluster(data, num_cluster, runs=10, level=2):
    km = KModes(n_clusters=num_cluster, init='Huang', n_init=runs, verbose=level)
    clusters = km.fit_predict(data)
    print(km.cluster_centroids_)
    return clusters
```

```python
cluster_dict = dict()
```

```python
for i in range(5,13):
    cluster_run = createcluster(data, i)
    cluster_dict[i] = cluster_run
```

```python
def saveAsPickle(name, data):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(data, f, pickle.HIGHEST_PROTOCOL)
    print("saved as pickle " + str(name))
```

```python
def saveAsJson(name, data):
    savname = name + '.json'
    with open(savname, 'w') as fp:
        json.dump(data,fp)
```
