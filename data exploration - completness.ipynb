```python
import pickle
import json
import pandas as pd
import numpy as np
```

```python
specjsons = dict()

with open('obj/specs.pkl', 'rb') as handle:
    specjsons = pickle.load(handle)

keys = list(specjsons.keys())
print(keys)
```

```python
print(specjsons[keys[2]].keys())
```

```python
key_dict = {}
key_list = list()
i = 0
for spec in specjsons:
    key_vector = []
    key_vector = list(specjsons[spec].keys())
    key_dict[spec] = key_vector
    key_list.append(key_vector)
    i += 1

print(key_list)

    
    
    
    
```

```python
with open('obj/'+ 'keysdict' + '.pkl', 'wb') as f:
        pickle.dump(key_dict, f, pickle.HIGHEST_PROTOCOL)
        
        
```

```python
with open('keys_spec.json', 'w') as fp:
        json.dump(key_dict,fp)
```

```python
spec_key_complete = []
spec_key_complete.append('swagger')
spec_key_complete.append('info')
spec_key_complete.append('host')
spec_key_complete.append('basePath')
spec_key_complete.append('schemes')
spec_key_complete.append('consumes')
spec_key_complete.append('produces')
spec_key_complete.append('paths')
spec_key_complete.append('definitions')
spec_key_complete.append('parameters')
spec_key_complete.append('responses')
spec_key_complete.append('securityDefinitions')
spec_key_complete.append('security')
spec_key_complete.append('tags')
spec_key_complete.append('externalDocs')









```

```python
print(spec_key_complete)
```

```python
len(spec_key_complete)
```

```python
onehotfields = dict()

for fields in key_dict:
    listfield = key_dict[fields]
    onehotvector = []
    for field in spec_key_complete:
        if field in listfield:
            onehotvector.append(1)
        else:
            onehotvector.append(0)
    print(len(onehotvector))
    onehotfields[fields] = onehotvector
```

```python
with open('obj/'+ 'onehotdict' + '.pkl', 'wb') as f:
        pickle.dump(onehotfields, f, pickle.HIGHEST_PROTOCOL)
```

```python
print(onehotfields[keys[1]])
```

```python
onehotdf = pd.DataFrame.from_dict(data=onehotfields, orient = 'index')
onehot = pd.DataFrame(onehotdf.values, index=keys,columns=spec_key_complete)




```

```python
onehot.to_csv('onehotfields.csv', encoding='utf8')
```

```python
onehot['implemented fields'] = onehot.eq(1).sum(axis=1)
onehot['missing fields'] = onehot.eq(0).sum(axis=1)

```

```python
onehot.to_csv('onehotfieldscount.csv', encoding='utf8')
```

```python
with open('obj/'+ 'onehotdf' + '.pkl', 'wb') as f:
        pickle.dump(onehot, f, pickle.HIGHEST_PROTOCOL)
```

```python
onehot["complete"] = (onehot["implemented fields"] + onehot["missing fields"])
```
